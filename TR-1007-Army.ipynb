```python
import subprocess

def execute(cmd):
    popen = subprocess.Popen(cmd, stdout=subprocess.PIPE, universal_newlines=True)
    for stdout_line in iter(popen.stdout.readline, ""):
        yield stdout_line 
    popen.stdout.close()
    return_code = popen.wait()
    if return_code:
        raise subprocess.CalledProcessError(return_code, cmd)

for TARGET_TICKER in LITERALLY_BAD_COMPANIES:
    #for out in execute(["python", ".\PVT-TR-1007.py", TARGET_TICKER]):
    #    print(out)
    for out in execute(["python", ".\PFC-TR-1007.py", TARGET_TICKER]):
        print(out)
    
```

```python
import pickle
from os import listdir
from os.path import isfile, join

base_path = "base-dir/"
onlyfiles = [f for f in listdir(base_path) if isfile(join(base_path, f))]

```

```python
import re, pickle, io 
stats = [p for p in onlyfiles if re.match('.*-stats', p)]
stat_imported = []
for p in stats:
    p = io.open(join(base_path,p), "rb")
    stat_imported += [pickle.load(p)]
    
```
